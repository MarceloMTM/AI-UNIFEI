```python
#Colocando a última coluna em um novo dataframe
df2 = df[['falha']]
df.drop('falha', axis=1, inplace=True)
df.head()
#df2.head()
```

```python
from scipy.stats import shapiro
```

```python
n_gauss= [''];
# Para cada coluna no DataFrame
for column in df.columns:
    # Realiza o teste de Shapiro-Wilk
    stat, p = shapiro(df[column])
    # Interpreta
    alpha = 0.05
    if p < alpha:
        print(f'Coluna {column}:')
        n_gauss.append(column);
        print('Estatística=%.3f, p=%.3f' % (stat, p))
        print('Amostra não parece Gaussiana (rejeitamos H0)\n')
    #else:
        #print('Amostra não parece Gaussiana (rejeitamos H0)\n')
```

```python
#lista de colunas que não são gaussianas
n_gauss = n_gauss[1:]
print(n_gauss)
```

```python
#criar um novo dataframe com base na lista de colunas que não são gaussianas
df_n_gauss = df[n_gauss]
df_n_gauss.head()
```

```python
#Removendo colunas não gaussianas do dataframe original
df.drop(n_gauss, axis=1, inplace=True)
df.head()
```

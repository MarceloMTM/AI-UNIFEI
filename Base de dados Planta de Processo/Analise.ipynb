```python
from scipy.stats import shapiro
```

```python
# Para cada coluna no DataFrame
for column in df.columns:
    # Realiza o teste de Shapiro-Wilk
    stat, p = shapiro(df[column])
    print(f'Coluna {column}:')
    print('Estatística=%.3f, p=%.3f' % (stat, p))
    # Interpreta
    alpha = 0.05
    if p > alpha:
        print('Amostra parece Gaussiana (não rejeitamos H0)\n')
    else:
        print('Amostra não parece Gaussiana (rejeitamos H0)\n')
```

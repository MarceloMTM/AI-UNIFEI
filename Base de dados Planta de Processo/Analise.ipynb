```python
#Colocando a Ãºltima coluna em um novo dataframe
df2 = df[['falha']]
df.drop('falha', axis=1, inplace=True)
df.head()
#df2.head()
```

```python
from scipy.stats import shapiro
```

```python
n_gauss= [''];
# Para cada coluna no DataFrame
for column in df.columns:
    # Realiza o teste de Shapiro-Wilk
    stat, p = shapiro(df[column])
    # Interpreta
    alpha = 0.05
    if p < alpha:
        print(f'Coluna {column}:')
        n_gauss.append(column);
        print('EstatÃ­stica=%.3f, p=%.3f' % (stat, p))
        print('Amostra nÃ£o parece Gaussiana (rejeitamos H0)\n')
    #else:
        #print('Amostra nÃ£o parece Gaussiana (rejeitamos H0)\n')
```

```python
#lista de colunas que nÃ£o sÃ£o gaussianas
n_gauss = n_gauss[1:]
print(n_gauss)
```

```python
#Construindo um Ãºnico grÃ¡fico com todos os histogramas usando df.hist() das colunas que nÃ£o sÃ£o gaussianas
df[n_gauss].hist(bins=20, figsize=(20,15))
plt.show()
```

```python
#criar um novo dataframe com base na lista de colunas que nÃ£o sÃ£o gaussianas
df_n_gauss = df[n_gauss]
df_n_gauss.head()
```

```python
#Removendo colunas nÃ£o gaussianas do dataframe original
df.drop(n_gauss, axis=1, inplace=True)
df.head()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X= df.drop('falha', axis=1).values
y= df['falha'].values

pca= PCA(n_components=2)
X_pca= pca.fit_transform(X)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.xlabel('First principal component')
plt.ylabel('Second principal component')
```

```python
from sklearn.ensemble import IsolationForest

clf = IsolationForest(random_state=42, contamination=0.2)
ano_y = clf.fit_predict(X)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=ano_y, cmap='viridis')

```
